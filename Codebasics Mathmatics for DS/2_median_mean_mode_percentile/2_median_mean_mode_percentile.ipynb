```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('income.csv')
df
```

```python
df.describe()
```

```python
df['Monthly Income ($)'].quantile(1, interpolation='linear')
df['Monthly Income ($)'].mean()
df['Monthly Income ($)'].median()
```

Remove outlier using quantile

```python
##Remove outlier using quantile

x = df['Monthly Income ($)'].quantile(.99)
df_no_outlier = df[df['Monthly Income ($)']<= x] ##this filters out those rows from df for which the boolean value is False ; 
                                          ##df['Monthly Income ($)']<= x  creates a Boolean series 

df_no_outlier
```

```python
df['Monthly Income ($)']<= x
```

Filling a missing value

```python
##Remove Sofia's income from the dataframe

df['Monthly Income ($)'][3] = np.NaN
df
```

```python
df= df.fillna(df['Monthly Income ($)'].median())
df
```

```python
##Remove outliers from Airbnb dataset

df_ex = pd.read_csv("AB_NYC_2019.csv")
df_ex
```

```python
df_ex.describe()
```

```python
percentile_10 = df_ex['price'].quantile(.1)


df_ex_new = df_ex[df_ex['price']>=percentile_10]

df_ex_new
```

```python
percentile_90 = df_ex['price'].quantile(.9)
percentile_10 = df_ex['price'].quantile(.1)
df_ex_new = df_ex[(df_ex['price']<percentile_90) & (df_ex['price']>=percentile_10)]
df_ex_new
```
